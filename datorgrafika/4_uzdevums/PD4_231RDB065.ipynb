Atvērt kādu brīvi izvēlētu attēlu, mērogojot attēlu tā, lai tā izmēri būtu platumā 500 pikseļi un augstumā 500 pikseļi (500,500) un iekrāsot atbilstošā attēla katru N-to pikseli uz X ass un katru M-to pikseli uz Y ass noteiktā krāsā<br>
Variants: 065 (x:6 y:10 sarkans)

```python
import numpy as np
from numpy import asarray

import cv2
from PIL import Image

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

fig = plt.figure(figsize=(10, 6), dpi=100, facecolor='w', edgecolor='k')
img = cv2.imread('photo.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ax1 = fig.add_subplot(1,3,1)
ax1.title.set_text("Ievades attēls")
ax1.imshow(img)
img = cv2.resize(img, (500,500))
ax2 = fig.add_subplot(1,3,2)
ax2.title.set_text("Mērogots attēls")
ax2.imshow(img)
for x in range(img.shape[0]):
    for y in range(img.shape[1]):
        if (x % 6) == 0 and (y % 10) == 0:
            img[y, x] = (255, 0, 0)
#plt.imshow(img)
ax3 = fig.add_subplot(1,3,3)
ax3.title.set_text("Punktots mērogots attēls")
ax3.imshow(img)
```
