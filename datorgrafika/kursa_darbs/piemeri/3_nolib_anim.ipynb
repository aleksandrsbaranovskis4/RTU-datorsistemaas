```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
import random

n_particles = 100

x = []
y = []
for _ in range(100):
    x.append(random.randint(0, 101))
    y.append(random.randint(0, 101))

for i in range(50):
    clear_output(wait=True)
    positions_x = x[i]
    positions_y = y[i]

    plt.plot(positions_x , positions_y,'ro')
    plt.axis((0 , 100.0 , 0 , 100.0))
    plt.show() 
    time.sleep(0.01)
```
